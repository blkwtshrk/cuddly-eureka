#Python

[Идентификаторы]
- Идентификаторы могут содержать буквы, цифры и подчеркивания (a-z, A-Z, 0-9, _), но не должны начинаться с цифры;
- Запрещено использовать специальные символы и ключевые слова языка;
- Лучше НЕ использовать буквы верхнего регистра в названиях, исключение: классы.



[Зарезервированные слова]

- Нельзя использовать ключевые слова как имена переменных, имена функций или какой-нибудь любой другой индефикатор;
- Они используются для определения синтаксиса и структуры в языке Python;
- Ключевые слова являются регистрозависимыми;
- Все слова - за исключением True, False и None - пишутся маленьким регистром.

| Категория                                | Ключевые слова                                |
| ---------------------------------------- | --------------------------------------------- |
| Логика и условия                         | `if`, `elif`, `else`                          |
| Логические операции                      | `and`, `or`, `not`                            |
| Циклы и управление                       | `for`, `while`, `break`, `continue`, `pass`   |
| Исключения                               | `try`, `except`, `finally`, `raise`, `assert` |
| Функции и возврат                        | `def`, `return`, `lambda`, `yield`            |
| Асинхронность                            | `async`, `await`                              |
| Классы и области видимости               | `class`, `global`, `nonlocal`                 |
| Импорты                                  | `import`, `from`, `as`                        |
| Контекстный менеджер                     | `with`                                        |
| Операторы принадлежности/тождественности | `in`, `is`                                    |
| Удаление                                 | `del`                                         |
| Литералы-синглтоны/булевы                | `True`, `False`, `None`                       |
