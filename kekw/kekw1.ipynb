# Python


## Идентификаторы

Правила для написания идентификаторов:
1. Комбинации букв в нижнем регистре (a - z)
2. Комбинация букв в верхнем регисре (A - Z)
3. Комбинация цифр (0 - 9) или _
4. Индентификатор НЕ может начинаться с цифры 1variable
5. Нельзя использовать спецсимволы !, @, #, $, %.





## Зарезервированные слова
- Мы не можем использовать ключевые слова как имена переменных, имена функций или какой-нибудь любой другой идентификатор. 
- Они используются, для определения синтаксиса и структуры в языке Pythone. 
- Ключевые слова регистрозависимые. Все слова за исключением True, False и None в маленьком регистре.

| Категория                                | Ключевые слова                                |
| ---------------------------------------- | --------------------------------------------- |
| Логика и условия                         | `if`, `elif`, `else`                          |
| Логические операции                      | `and`, `or`, `not`                            |
| Циклы и управление                       | `for`, `while`, `break`, `continue`, `pass`   |
| Исключения                               | `try`, `except`, `finally`, `raise`, `assert` |
| Функции и возврат                        | `def`, `return`, `lambda`, `yield`            |
| Асинхронность                            | `async`, `await`                              |
| Классы и области видимости               | `class`, `global`, `nonlocal`                 |
| Импорты                                  | `import`, `from`, `as`                        |
| Контекстный менеджер                     | `with`                                        |
| Операторы принадлежности/тождественности | `in`, `is`                                    |
| Удаление                                 | `del`                                         |
| Литералы-синглтоны/булевы                | `True`, `False`, `None`                       |

Python - регистрозависимый язык. Это значит, что 'Kekwait' != 'kekwait' это не одно и тоже.

## Переменная
- Именованный участок памяти, обладающий некоторым типом.
- К примеру, variable = 2
- Переменная - это элемент, который хранит в себе какие либо данные. Переменная никогда не бывает абсолютно пустой.
- Но может храниться пустая строка
- К примеру, empty = ' '


### Объявление переменной
- Переменные в Python не типизированные. Это значит, что при создании переменной не требуется указывать какой тип данных будет в ней храниться.
- Переменную не нужно объявлять заранее, до присвоения ей каких - либо данных.
- Объявление переменной происходит простым присвоением ей значение. 'Название' = 'данные'
- Переменная фиксированна, но значине данных переменной изменяемо

```python
a=12
print(a+12*4)
```

### Операторы
| Тип оператора                | Операторы                                       | Назначение                        | Пример             |
| ---------------------------- | ----------------------------------------------- | --------------------------------- | ------------------ |
| **Арифметические**           | `+`, `-`, `*`, `/`, `//`, `%`, `**`             | Математические вычисления         | `5 + 3`, `10 // 3` |
| **Операторы присваивания**   | `=`, `+=`, `-=`, `*=`, `/=`, `//=`, `%=`, `**=` | Присваивание и изменение значения | `x += 5`           |
| **Операторы сравнения**      | `==`, `!=`, `>`, `<`, `>=`, `<=`                | Сравнение значений                | `a > b`            |
| **Логические**               | `and`, `or`, `not`                              | Логические операции               | `a > 5 and b < 10` |
| **Побитовые**                | `&`, `\|`, `^`, `~`, `<<`, `>>`                 | Операции над битами               | `5 & 3`            |
| **Операторы принадлежности** | `in`, `not in`                                  | Проверка наличия элемента         | `'a' in "cat"`     |
| **Операторы идентичности**   | `is`, `is not`                                  | Проверка, что объекты одинаковые  | `x is None`        |

### Порядок операций
1. Parentheses - Скобки
2. Exponentiation - Возведение в степень
3. Multiplication - Умножение
4. Division - Деление
5. Addition - Сложение
6. Subtraction - Вычитание
